```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
     ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Purchase_Amount'].sum().sort_values(by='Purchase_Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Purchase_Amount'].sum().sort_values(by='Purchase_Amount', ascending=False)
sns.barplot(x = 'Gender' ,y= 'Purchase_Amount' ,data = sales_gen)
```

```python
df.groupby(['Gender'], as_index=False)['Rating'].sum().sort_values(by='Rating', ascending=False)
```

```python
sales_gen2 = df.groupby(['Gender'], as_index=False)['Rating'].sum().sort_values(by='Rating', ascending=False)
sns.barplot(x = 'Gender' ,y= 'Rating' ,data = sales_gen2)
```

```python
sns.pairplot(df, hue='Gender')
plt.show()
```

```python
g = sns.pairplot(df, hue="Gender", diag_kind="hist")
```

```python

sales_gen8 = df.groupby(['Product_Name'], as_index=False)['Purchase_Amount'].sum()
sales_gen8 = sales_gen8.sort_values(by='Purchase_Amount', ascending=False)

plt.figure(figsize=(12, 6)) 
sns.barplot(x='Product_Name', y='Purchase_Amount', data=sales_gen8)

plt.xticks(rotation=45) 
plt.title('Total Purchase Amount by Product')
plt.xlabel('Product Name')
plt.ylabel('Total Purchase Amount')
plt.tight_layout()
plt.show()
```

```python
sales_gen2 = df.groupby(['Category'], as_index=False)['Purchase_Amount'].sum().sort_values(by='Purchase_Amount', ascending=False)
sns.histplot(x = 'Category' ,y= 'Purchase_Amount' ,data = sales_gen2)
```

```python
sales_gen4 = df.groupby(['Gender'], as_index=False)['Age'].count().sort_values(by='Age', ascending=False)
sns.histplot(x = 'Gender' ,y= 'Age' ,data = sales_gen4)

```
